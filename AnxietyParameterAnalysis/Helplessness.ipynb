```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define parameters
anxiety_strength = 0.05
depression_strength = 0.1
time_steps = 300
```

```python
anxiety = np.zeros(time_steps)
depression = np.zeros(time_steps)
```

```python
# Initial conditions
anxiety[0] = 0.5
depression[0] = 0.0
```

```python
for t in range(1, time_steps):
    anxiety[t] = anxiety[t-1] + anxiety_strength - depression_strength * anxiety[t-1]
    depression[t] = depression[t-1] + depression_strength * anxiety[t-1]
```

```python
# Plotting
plt.figure(figsize=(20, 20))
plt.plot(anxiety, label='Anxiety')
plt.plot(depression, label='Depression')
plt.xlabel('Time')
plt.ylabel('Intensity')
plt.title('Transition from Anxiety to Depression')
plt.legend()
plt.grid(True)
plt.show()
```
